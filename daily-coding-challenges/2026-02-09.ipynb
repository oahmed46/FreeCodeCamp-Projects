2026 Winter Games Day 4: Ski Jumping
Given distance points, style points, a wind compensation value, and K-point bonus value, calculate your score for the ski jump and determine if you won a medal or not.

Your score is calculated by summing the above four values.
The current total scores of the other jumpers are:

165.5
172.0
158.0
180.0
169.5
175.0
162.0
170.0
If your score is the best, return "Gold"
If it's second best, return "Silver"
If it's third best, return "Bronze"
Otherwise, return "No Medal"

```python
def ski_jump_medal(distance_points, style_points, wind_comp, k_point_bonus):

    # Code assumes list is unordered
    total_scores = [165.5, 172.0, 158.0, 180.0, 169.5, 175.0, 162.0, 170.0]
    
    # Code to order list
    # total_scores = sorted(total_scores)[::-1]
    
    print(f"jumper total scores: {total_scores}")

    your_score = distance_points + style_points + wind_comp + k_point_bonus
    print(f"your total score: {your_score}")

    ranking_score = 0
    for jumper_score in total_scores:
        print(f"jumper_score {jumper_score} | ranking_score: {ranking_score}")
        if jumper_score > your_score:
            ranking_score += 1

    if ranking_score == 0:
        return "Gold"
    elif ranking_score == 1:
        return "Silver"
    elif ranking_score == 2:
        return "Bronze"
    else:
        return "No Medal"
```

```python
ski_jump_medal(125.0, 58.0, 0.0, 6.0) # "Gold"
```

```python
ski_jump_medal(119.0, 50.0, 1.0, 4.0) # "Bronze"
```

```python
ski_jump_medal(122.0, 52.0, -1.0, 4.0) # "Silver"
```

```python
ski_jump_medal(118.0, 50.5, -1.5, 4.0) # "No Medal"
```

```python
ski_jump_medal(124.0, 50.5, 2.0, 5.0) # "Gold"
```
