2026 Winter Games Day 5: Cross-Country Skiing
Given an array of finish times for a cross-country ski race, convert them into times behind the winner.

Given times are strings in "H:MM:SS" format.
Given times will be in order from fastest to slowest.
The winners time (fastest time) should correspond to "0".
Each other time should show the time behind the winner, in the format "+M:SS".
For example, given ["1:25:32", "1:26:10", "1:27:05"], return ["0", "+0:38", "+1:33"].

```python
def get_relative_results(results):
    
    # initial logic attempt
    # relative_results = ["0"]
    # for time in results[1:]:

    #     minutes = time[2:4] - results[0][2:4]
    #     seconds = time[5:7] - results[0][5:7]

    #     if minutes > 0:
    #         total_time = 60 * minutes + time[5:7] - results[0][5:7]
    #         if total_time/60 < 1:
    #             seconds = total_time
    #         else:
    #             minutes = total_time//60
    #             seconds = total_time%60

    #     relative_results.append(f"+{minutes}:{seconds}")

    # return relative_results
    
    # second logic attempt
    results_in_seconds = [int(time[0])*60*60 + int(time[2:4])*60 + int(time[5:7]) for time in results]

    time_behind_winner_in_seconds = [time - results_in_seconds[0] for time in results_in_seconds]

    time_behind_winner = ["0"]
    for time_behind in time_behind_winner_in_seconds[1:]:
        time_behind_winner.append(f"+{time_behind//60}:{time_behind%60:02d}")

    return time_behind_winner
```

```python
get_relative_results(["1:25:32", "1:26:10", "1:27:05"]) #["0", "+0:38", "+1:33"]
```
