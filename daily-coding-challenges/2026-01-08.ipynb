Sorted Array?
Given an array of numbers, determine if the numbers are sorted in ascending order, descending order, or neither.

If the given array is:

In ascending order (lowest to highest), return "Ascending".
In descending order (highest to lowest), return "Descending".
Not sorted in ascending or descending order, return "Not sorted".

```python
# Initial implementation
# def is_sorted(arr):
#     ascending = True
#     descending = True
#     for i in range(len(arr)-1):
#         if arr[i] < arr[i+1]:
#             descending = False 
#         if arr[i] > arr[i+1]:
#             ascending = False
#     if ascending and not descending:
#         return "Ascending"
#     if not ascending and descending:
#         return "Descending"
#     else:
#         return "Not sorted"

# Improved implementation    
def is_sorted(arr):
    # Prevents edge cases such as [] or [5]
    if len(arr) < 2:
        return "Ascending"
    
    # Flags for ascending and descending
    ascending_possible = True
    descending_possible = True

    # Ensures we do not look outside the array
    for i in range(len(arr)-1):
        if arr[i] < arr[i+1]:
            descending_possible = False 
        elif arr[i] > arr[i+1]:
            ascending_possible = False

        # Early check to prevent continuing through the loop
        if not ascending_possible and not descending_possible:
            return "Not sorted"

    if ascending_possible:
        return "Ascending"
    if descending_possible:
        return "Descending"
    return "Not sorted"
```

```python
is_sorted([1, 2, 3, 4, 5]) # "Ascending"
```

```python
is_sorted([10, 8, 6, 4, 2]) # "Descending"
```

```python
is_sorted([1, 3, 2, 4, 5]) # "Not sorted"
```

```python
is_sorted([314, 271, 161, 0.57]) # "Descending"
```
